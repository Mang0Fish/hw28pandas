```python
import pandas as pd
import numpy as np

#1
df = pd.DataFrame({
'CustomerID': np.arange(1, 11),
'FirstName': ['John', 'Steve', 'Anna', 'Mike', 'Emily', 'Jake', 'Laura', 'Nick', 'Olivia',
'David'],
'LastName': ['Smith', 'Johnson', 'Williams', 'Jones', 'Brown', 'Davis', 'Miller',
'Wilson', 'Moore', 'Taylor'],
'Age': np.random.randint(25, 65, size=10),
'TotalPurchaseAmount': np.random.randint(100, 1000, size=10)
})


#2
df['classification'] = df['TotalPurchaseAmount'].apply(lambda x: 'Low' if x < 500 else 'High')

#3 the question is wrong cause age is from 25 to 65 and you said bigger or smaller than 20, so Ill choose 45 instead
df['classification_by_age'] = np.vectorize(lambda classi, age: 'Low_Young' if classi == 'Low' and age < 45 else 'High_Old' if classi == 'High' and age >= 45
else 'Low_old' if classi == 'Low' and age >= 45 else 'High_Young')(df['classification'], df['Age'])
df
```
